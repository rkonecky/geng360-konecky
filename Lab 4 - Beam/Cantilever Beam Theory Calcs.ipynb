## Displacement

$$\delta_{max} = \frac{FL^3}{3EI}$$

$$I = \frac{bh^3}{12}$$

```python
import numpy as np # You only need to import this in the first cell

F = 0.110  # 0.5  # lb, Applied Force
L = 12.276  # 12.25  # in, Beam Length
E = 1e7  # psi, Young's Modulus for 6061 Aluminum
rho = 0.09754  # lb/in^3, density
b = 0.5  # in, beam width
h = 0.125  # in, beam height

I = b * h**3 / 12  # in^4, moment of inertia ; recall that in python ** is how you raise something to a power (not ^)

delta = F * L**3 / (3 * E * I)

# f-string: {delta:.3g} shows delta with 3 significant figures (general format)
# other options: {delta:.3f} fixed 3 decimals, {delta:.2e} scientific, {delta:.0f} nearest int

print(f'Displacement is {delta:.3g} in') 
```
